```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# import csv for Citibike 2014 usage data
df_2014 = pd.read_csv('resources/2014_citibike_data.csv')
df_2014.head()
```

```python
# remove lat/lng columns and renamed remaining columns
new_df_2014 = df_2014.drop(['start station longitude', 'start station latitude', 'end station longitude', 'end station latitude'], axis=1)
new_df_2014.rename(columns={'tripduration': 'Trip_Duration'}, inplace=True)
new_df_2014.rename(columns={'starttime': 'Start_Time'}, inplace=True)
new_df_2014.rename(columns={'stoptime': 'Stop_Time'}, inplace=True)
new_df_2014.rename(columns={'bikeid': 'Bike_ID'}, inplace=True)
new_df_2014.rename(columns={'usertype': 'User_Type'}, inplace=True)
new_df_2014.rename(columns={'birth year': 'Birth_Year'}, inplace=True)
new_df_2014.rename(columns={'gender': 'Gender'}, inplace=True)
new_df_2014.rename(columns={'start station name': 'Start_Station_Name'}, inplace=True)
new_df_2014.rename(columns={'end station name': 'End_Station_Name'}, inplace=True)
new_df_2014.rename(columns={'start station id': 'Start_Station_ID'}, inplace=True)
new_df_2014.rename(columns={'end station id': 'End_Station_ID'}, inplace=True)
new_df_2014.head(15)
```

```python
# replace codes in Gender field to a more easily indentified name
new_df_2014['Gender'].replace(1, 'Male', inplace=True)
new_df_2014['Gender'].replace(2, 'Female', inplace=True)
new_df_2014['Gender'].replace(0, 'Unknown', inplace=True)
new_df_2014.head(15)
```

```python
# replace any '\N' values with empty strings, then convert to NaN values
new_df_2014 = new_df_2014.replace(r'\\N','', regex=True)
new_df_2014['Birth_Year'].replace('', pd.np.nan, inplace=True)
new_df_2014.head(15)
```

```python
# get initial record count
new_df_2014.count()
```

```python
# drop any rows with NAN values
new_df_2014.dropna(how="any", inplace=True)
new_df_2014.head(15)
```

```python
# get final record count after cleaning
new_df_2014.count()
```

```python
# rename dataframe to finalize it before pushing to SQL database
final_df_2014 = new_df_2014
final_df_2014.head()
```

```python
# create SQL connection
connection = "root:rootroot@localhost:3306/citibike_db"
engine = create_engine(f'mysql://{connection}')
```

```python
# confirm empty db 
engine.table_names()
```

```python
# create table for 2014 data
final_df_2014.to_sql(name='citibike_2014', con=engine, if_exists='append', index=False)
```

```python
# query SQL db to make sure the data is present
pd.read_sql_query('select * from citibike_2014', con=engine).head(15)
```
