```python
import pandas as pd
import numpy as np
```

```python
df_2018 = pd.read_csv('resources/2018_CitiBike_data.csv')
df_2018.head(10)
```

```python
newdf2018 = df_2018.drop(['start station longitude', 'start station latitude', 'end station longitude', 'end station latitude'], axis=1)
newdf2018.rename(columns={'tripduration': 'Trip_Duration'}, inplace=True)
newdf2018.rename(columns={'starttime': 'Start_Time'}, inplace=True)
newdf2018.rename(columns={'stoptime': 'Stop_Time'}, inplace=True)
newdf2018.rename(columns={'bikeid': 'Bike_ID'}, inplace=True)
newdf2018.rename(columns={'usertype': 'User_Type'}, inplace=True)
newdf2018.rename(columns={'birth year': 'Birth_Year'}, inplace=True)
newdf2018.rename(columns={'gender': 'Gender'}, inplace=True)
newdf2018.rename(columns={'start station name': 'Start_Station_Name'}, inplace=True)
newdf2018.rename(columns={'end station name': 'End_Station_Name'}, inplace=True)
newdf2018.rename(columns={'start station id': 'Start_Station_ID'}, inplace=True)
newdf2018.rename(columns={'end station id': 'Start_Station_ID'}, inplace=True)
newdf2018.head(10)
```

```python
newdf2018.dropna(how='all')
newdf2018.head(10)
```

```python
newdf2018 = newdf2018.replace(r'\\N','', regex=True)
newdf2018['Birth_Year'].replace('', pd.np.nan, inplace=True)
newdf2018.head(15)
```
